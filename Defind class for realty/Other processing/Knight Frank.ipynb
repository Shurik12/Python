### Common functions for all three classes

```python
def get_address(file, i, j, n = 20):
    div = browser.find_elements_by_xpath('//div[@class="listing__cards"]/div[%s]/div[2]/div[2]/div[1]/div[1]/div[1]/span' % str(i))
    if (len(div) > 0):
        span = div[-1]
    else:
        span = browser.find_elements_by_xpath('//div[@class="listing__cards"]/div[%s]/div[2]/div[2]/div[1]/span' % str(i))[-1]
    href = browser.find_elements_by_xpath('//div[@class="listing__cards"]/div[%s]/div[2]/div[2]/div[1]/div[1]/div[1]/a' % str(i))
    if (len(href) > 0):
        a_s = href
    else:
        a_s = browser.find_elements_by_xpath('//div[@class="listing__cards"]/div[%s]/div[2]/div[2]/div[1]/a' % str(i))
    if i < 3:
        line = str(i+(j-1)*n) + "\t" + span.get_attribute("innerText")
    else:
        line = str(i-1+(j-1)*n) + "\t" + span.get_attribute("innerText")
    for k in range(len(a_s)):
        line = line + "\t" + a_s[k].get_attribute("text")
    print(line)
    line = line + "\n"
    file.write(line)

def get_address_1(file, i, j, n = 20):
    div = browser.find_elements_by_xpath('//div[@class="listing__cards"]/div[%s]/div[2]/div[2]/div[1]/div[1]/div[1]/span' % str(i))
    if (len(div) > 0):
        span = div[-1]
    else:
        span = browser.find_elements_by_xpath('//div[@class="listing__cards"]/div[%s]/div[2]/div[2]/div[1]/span' % str(i))[-1]
    href = browser.find_elements_by_xpath('//div[@class="listing__cards"]/div[%s]/div[2]/div[2]/div[1]/div[1]/div[1]/a' % str(i))
    if (len(href) > 0):
        a_s = href
    else:
        a_s = browser.find_elements_by_xpath('//div[@class="listing__cards"]/div[%s]/div[2]/div[2]/div[1]/a' % str(i))
    line = str(i+(j-1)*n) + "\t" + span.get_attribute("innerText")
    for k in range(len(a_s)):
        line = line + "\t" + a_s[k].get_attribute("text")
    print(line)
    line = line + "\n"
    file.write(line)

def get_image(i, j, n = 20, time_sleep = 3):
    elem = browser.find_element_by_xpath('//div[@class="listing__cards"]/div[%s]/a' % str(i))
    href = elem.get_attribute('href')
    browser.get(href)
    time.sleep(time_sleep)
    img = browser.find_element_by_tag_name("img")
    src = img.get_attribute('src')
    if i < 3:
        urllib.request.urlretrieve(src, "Class A/%s.jpg" % str(i+(j-1)*n))
    else: 
        urllib.request.urlretrieve(src, "Class A/%s.jpg" % str(i-1+(j-1)*n))
    browser.back()
    time.sleep(time_sleep - 1)
    
def get_image_1(i, j, n = 20, time_sleep = 2):
    elem = browser.find_element_by_xpath('//div[@class="listing__cards"]/div[%s]/a' % str(i))
    href = elem.get_attribute('href')
    browser.get(href)
    time.sleep(time_sleep)
    img = browser.find_element_by_tag_name("img")
    src = img.get_attribute('src')
    urllib.request.urlretrieve(src, "Class A/%s.jpg" % str(i+(j-1)*n))
    browser.back()
    time.sleep(time_sleep - 1)
    
```

```python
# Парсим изображения здания класса A

def processing_class_A():
    try:
        # open file for writing
        file = open("addresses_class_A.tsv", "w")

        for j in range(1, 20):
            print("This is page "+str(j))
            if j % 2 == 1:
                for i in range(1, 42):
                    print("This is "+ str(i) +" iteration")
                    if i == 3:
                        continue
                    get_address(file, i, j, 40)
                    #get_image(i, j, 40, 2)

                elem = browser.find_element_by_class_name("pagination__btn--next")
                href = elem.get_attribute('href')
                browser.get(href)
            else:
                for i in range(1, 41):
                    print("This is "+ str(i) +" iteration")
                    get_address_1(file, i, j, 40)
                    #get_image_1(i, j, 40, 2)

                elem = browser.find_element_by_class_name("pagination__btn--next")
                href = elem.get_attribute('href')
                browser.get(href)
                time.sleep(2)
    except:
        print ("Done!")
    finally:
        file.close()
```
