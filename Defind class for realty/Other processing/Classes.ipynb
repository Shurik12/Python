```python
with open("Class2.tsv", "w") as f_w:
    with open("Class1.tsv", "r") as f_r:
        f_w.write("id\tprice\tstreet\thouse\tkorpus\tstroenie\tbuldingclass\tbuildyear\tfloorscount\n")
        for line in f_r:
            line = line.split('\t')
            if 'c' in line[6]:
                line[6] = 'c'
            elif 'b' in line[6]:
                line[6] = 'b'
            elif 'a' in line[6]:
                line[6] = 'a'
            else:
                line[6] = "None"
            line_w = ''
            for word in line:
                line_w += word + '\t'
            line_w = line_w.strip() + '\n'
            f_w.write(line_w)
```

```python
def get_class_a():
    class_a = []
    with open("Class_A_1.tsv", "r") as f_r:
        for line in f_r:
            line = line.strip('\n').split('\t')
            class_a.append(line)
    return class_a

def check_is_c(line):
    price = line[1]
    buildyear = line[7]
    floorscount = line[8]
    buildingclass = line[6]
    if price != "None":
        if float(price) <= 18000:
            return 1
    if "c" in buildingclass:
        return 1
    if floorscount.strip() != "None":
        if int(floorscount) <= 4:
            return 1
    if buildyear.strip() != "None":    
        if int(buildyear) > 1000 and int(buildyear) <= 2000:
            return 1
    return 0

def check_is_b(line):
    price = line[1].strip()
    buildyear = line[7].strip()
    floorscount = line[8].strip()
    buildingclass = line[6].strip()
    if price != "None":
        if float(price) <= 35000:
            return 1
    if "b" in buildingclass:
        return 1
    if floorscount.strip() != "None":
        if int(floorscount) <= 7:
            return 1
    if buildyear.strip() != "None":
        if int(buildyear) > 1000 and int(buildyear) <= 2010:
            return 1
    return 0
```

```python
def get_result(class_a):
    is_c = 0
    is_b = 0
    is_a = 0
    total = 0
    with open("Result.tsv", "w") as f_w:
        f_w.write("id\tstreet\thouse\tkorpus\tstroenie\tbuldingclass\n")
        with open("Class2.tsv", "r") as f_r:
            line = f_r.readline()
            line = f_r.readline()
            while line:
                total += 1
                line = line.strip('\n').split('\t')
                if line[2:6] in class_a:
                    is_a += 1
                    line_w = line[0]+'\t'+line[2]+'\t'+line[3]+'\t'+line[4]+'\t'+line[5]+'\ta\n'
                    print (line[2:6])
                    f_w.write(line_w)
                elif check_is_c(line):
                    is_c += 1 
                    line_w = line[0]+'\t'+line[2]+'\t'+line[3]+'\t'+line[4]+'\t'+line[5]+'\tc\n'
                    f_w.write(line_w)
                elif check_is_b(line):
                    is_b += 1
                    line_w = line[0]+'\t'+line[2]+'\t'+line[3]+'\t'+line[4]+'\t'+line[5]+'\tb\n'
                    f_w.write(line_w)
                else:
                    line_w = line[0]+'\t'+line[2]+'\t'+line[3]+'\t'+line[4]+'\t'+line[5]+'\tNone\n'
                    f_w.write(line_w)
                line = f_r.readline()
    return total, is_a, is_b, is_c 
```

```python
class_a = get_class_a()
result = get_result(class_a)
```
