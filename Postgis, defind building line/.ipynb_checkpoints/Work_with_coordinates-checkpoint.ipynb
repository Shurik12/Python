```python
sql1 = """
        SELECT street
        FROM moscow_buildings 
        GROUP BY street
        """
sql2 = """
        SELECT name 
        FROM streets
        """
sql3 = """
        SELECT street, house, ST_AsEWKT(polygon)
        FROM moscow_buildings
        """
sql4 = """
        SELECT name, ST_AsEWKT(way)
        FROM streets
        LIMIT 1
        """
sql5 = """
        SELECT ST_AsEWKT(way) 
        FROM streets
        WHERE name = '%s'
        """
sql6 = """
        SELECT ST_AsText(ST_ClosestPoint(way, ST_Geomfromtext('%s', 4326)))
        FROM streets
        WHERE name = '%s'
        """
sql7 = """
        SELECT count(*)
        FROM moscow_buildings
        WHERE street = '%s' and house <> '%s' and 
                ST_Intersects(ST_LineFromText('%s', 4326), polygon)
        """
```

```python
# получаем улицы изтаблиц moscow_buildings и streets
def get_streets1_streets2(cursor):
    streets1 = []
    streets2 = []
    cursor.execute(sql1)
    for line in cursor.fetchall():
        streets1.append(line[0])
    cursor.execute(sql2)
    for line in cursor.fetchall():
        streets2.append(line[0])
    return streets1, streets2

# получить улицы, входищие в moscow_buildings и не входящие в streets
def get_streets3(streets1, streets2):
    count = 0
    streets3 = []
    for street in streets1:
        if street in streets2: 
            count += 1
        else:
            streets3.append(street)
    return count, streets3

# создать словарь улиц по streets1
def get_dict_streets1(streets1):
    dict_streets = {}
    for line in streets1:
        dict_streets[line] = {}
    return dict_streets

# заполнить домами и их координатами dict_streets
def fill_dict_streets(cursor, dict_streets):
    cursor.execute(sql3)
    for line in cursor.fetchall():
        coords = line[2][line[2].find("(")+2:line[2].find(")")].split(",")
        dict_streets[line[0]][line[1]] = {} 
        for i in range(len(coords)):
            dict_streets[line[0]][line[1]]["point%s" % str(i+1)] = coords[i]
            
# заполнить домами и их координатами dict_streets
def fill_dict_streets_1(cursor, dict_streets):
    cursor.execute(sql3)
    for line in cursor.fetchall():
        coords = line[2][line[2].find("(")+2:line[2].find(")")].split(",")
        coords = processing_coords(coords)
        dict_streets[line[0]][line[1]] = coords
    return dict_streets

def length(x1, y1, x2, y2, x0, y0):
    A = y2 - y1
    B = -(x2 - x1)
    C = y1*x2 - x1*y2
    return abs(A*x0 + B*y0 + C) / math.sqrt(A*A + B*B)
def get_distance(x1, y1, x2, y2):
    dx = x2-x1
    dy = y2-y1
    return math.sqrt(dx*dx+dy*dy)
    

def get_dict_streets2(cursor, dict_streets):
    dict_streets2 = {}
    for street in dict_streets.keys():
        cursor.execute(sql2 % street)
        batch = cursor.fetchall()
        if batch:
            dict_streets2[street] = {}
            i = 0
            for line in batch:
                i += 1
                line = line[0]
                coords = line[line.find("(")+1:line.find(")")].split(",")
                dict_streets2[street]["part%s" % str(i)] = coords
    return dict_streets2

# получить проекцию точки на улицу
def get_projection(point_str, street):
    x2, y2 = map(float, point_str.split(" "))
    params = ("POINT(%s %s)" % (x2, y2), street)
    cursor.execute(sql6 % params)
    distance = -1
    projection = ""
    for line in cursor.fetchall():
        line = line[0]
        point = line[line.find("(")+1:line.find(")")]
        x1, y1 = map(float, point.split(" "))
        new_distance = dist(x1, y1, x2, y2)
        if new_distance < distance or distance < 0:
            distance = new_distance
            projection = point
    return distance, projection

# получить словарь проекций
def get_dict_projections(dict_streets):
    dict_projections = {}
    for street in dict_streets.keys():
        dict_projections[street] = {}
        for house in dict_streets[street].keys():
            dict_projections[street][house] = {}
            for point in dict_streets[street][house].keys():
                distance, projection = get_projection(dict_streets[street][house][point], street)
                dict_projections[street][house][point] = {}
                dict_projections[street][house][point]["distance"] = distance
                dict_projections[street][house][point]["projection"] = projection
    return dict_projections

# получить словарь проекций
def get_dict_projections_1(dict_streets):
    dict_projections = {}
    for street in dict_streets.keys():
        dict_projections[street] = {}
        for house in dict_streets[street].keys():
            dict_projections[street][house] = {}
            for point in dict_streets[street][house]:
                distance, projection = get_projection(point, street)
                dict_projections[street][house][point] = projection
    return dict_projections

# вставить улицу в таблицу (проекции точек домов на улицу)
def insert_street(dict_projections, dict_streets, cursor):
    for street in dict_projections.keys():
        if street == "Литовский бульвар":
            for house in dict_projections[street].keys():
                for point in dict_projections[street][house].keys():
                    #print (street, house, point, dict_projections[street][house][point]["projection"])
                    point1 = dict_projections[street][house][point]["projection"]
                    point2 = dict_streets[street][house][point]
                    params = (street, house, 'LINESTRING(%s, %s)' % (point1, point2))
                    cursor.execute(sql % params)
            break
    connection.commit()
    
# определить линию застройки на примере Литов
def difind_building_line(dict_projections, dict_streets, cursor):
    for street in dict_projections.keys():
        if street == "Литовский бульвар":
            for house in dict_projections[street].keys():
                building_line = 1
                for point in dict_projections[street][house].keys():
                    #print (street, house, point, dict_projections[street][house][point]["projection"])
                    point1 = dict_projections[street][house][point]["projection"]
                    point2 = dict_streets[street][house][point]
                    params = (street, house, 'LINESTRING(%s, %s)' % (point1, point2))
                    #params = (street, house)
                    cursor.execute(sql7 % params)
                    if cursor.fetchall()[0][0] != 0:
                        building_line = 2
                        break
                print (house, building_line)
            break
            
def processing_coords(coords):
    min_lon, max_lon = 100, 0
    min_lat, max_lat = 100, 0
    for i in range(len(coords)):
        lon, lat = map(float, coords[i].split(" "))
        if min_lon > lon: min_lon = lon
        if max_lon < lon: max_lon = lon
        if min_lat > lat: min_lat = lat
        if max_lat < lat: max_lat = lat
    d_lon = (max_lon - min_lon) / 6
    d_lat = (max_lat - min_lat) / 6
    point = []
    point.append(str(min_lon + d_lon) + " " + str(min_lat + d_lat))
    point.append(str(max_lon - d_lon) + " " + str(min_lat + d_lat))
    point.append(str(min_lon + d_lon) + " " + str(max_lat - d_lat))
    point.append(str(max_lon - d_lon) + " " + str(max_lat - d_lat))
    return point
```
